```python
import numpy as np
import pandas as pd
```

```python
yien1 = pd.read_csv('../1-Source_Data/yien1.csv') 
```

```python
# 1.合并数据

# 豆瓣数据
data1 = pd.read_csv('../1-Source_Data/douban.csv')

# 合并艺恩数据
yien1 = pd.read_csv('../1-Source_Data/yien1.csv')  
yien2 = pd.read_csv('../1-Source_Data/yien2.csv') 
data2 = pd.concat([yien1,yien2])

# 处理艺恩票房数据
data2 = data2.loc[:,data2.columns != '_id']
data2 = data2.loc[:,['MovieName','SumBoxOffice','MovieFxAll','MovieZz']]
data2.columns = ['电影名','票房','发行公司','制片公司']

#合并豆瓣艺恩数据
data = data1.merge(data2,left_on="电影名",right_on="电影名",how="outer")
```

```python
# 2.电影名重名处理
data.drop_duplicates(subset=['电影名'],inplace=True)

print()

# 查看处理后情况
data.info()
```

```python
# 3.电影导演处理

#查看导演为空的数据情况
data.dropna(subset=['导演']).info()

#删除导演为空的数据
data.dropna(subset=['导演'],inplace=True)

#删除暂无导演信息的数据
data = data[data['导演'] != '暂无导演信息']

print()

# 查看处理后情况
data.info()
```

```python
# 4.电影编剧处理

#查看编剧为空的数据情况
data.dropna(subset=['编剧']).info()

#删除编剧为空的数据
data.dropna(subset=['编剧'],inplace=True)

#删除暂无编剧信息的数据
data = data[data['编剧'] != '暂无编剧信息']

print()

# 查看处理后情况
data.info()
```

```python
# 5.电影演员处理

#查看演员为空的数据情况(数据匹配度提高，所以删除)
data.dropna(subset=['演员']).info()

#删除演员为空的数据
data.dropna(subset=['演员'],inplace=True)

#删除暂无演员信息的数据
data = data[data['演员'] != '暂无演员信息']

print()

# 查看处理后情况
data.info()
```

```python
# 6.上映时间处理

# 查看上映时间的数据情况
data.dropna(subset=['上映时间']).info()

# 删除上映时间为空的数据
data.dropna(subset=['上映时间'],inplace=True)

# 删除暂无上映时间的数据
data = data[data['上映时间'] != '暂无上映时间']

# 提取上映时间年份
# (1).将上映时间这列数据取出将ndarray转为list
list_showtime = data.loc[:,'上映时间'].values.tolist()

# (2).取出年份
lists_showtime = []
for list in list_showtime:
    lists_showtime.append(list[:4])
    
# (3).拼接处理后的数据
data_showtime = pd.DataFrame(lists_showtime,index=data.index,columns=['上映时间(年)'])
datas = data.merge(data_showtime,left_on=data.index,right_on=data_showtime.index,how="outer")

# (4).将原上映时间和key_0列删除
columns = datas.columns[(datas.columns!='key_0') & (datas.columns!='上映时间')]
data = datas.loc[:,columns]

print()

# 查看处理后情况
data.info()
```

```python
# 7.电影片长处理

# 查看片长不存在的数据情况
data[data.loc[:,'片长']!='暂无电影时长'].info()

# 将暂无电影时长数据改为0分钟，但类型为str
data.loc[data[(data['片长']=='暂无电影时长')].index,'片长']='0分钟'

# 将空数据替换为0分钟
data['片长'].fillna('0分钟',inplace=True)

# 用正则表达式处理电影片长只剩数值部分

# (1).将片长这列数据取出将ndarray转为list
list_time = data.loc[:,'片长'].values.tolist()

# (2).正则表达式处理
lists = []
import re
for list in list_time:
    lists.append(re.sub("\D", "", list))

# (3).处理特殊数据    
lists_finally = []
for i in lists:
    if len(i) < 3:
        lists_finally.append(i)
    elif len(i) > 3 and int(i[:2]) < 20:
        lists_finally.append(i[:3])
    elif len(i) > 3 and int(i[:2]) > 20:
        lists_finally.append(i[:2])
    else:
        lists_finally.append(i)

# (4).将0替换为120,因为电影时长一般都为120
lists_end = []
for list in lists_finally:
    if list == str(0):
        lists_end.append(120)
    else:
        lists_end.append(list)
        
# (5).拼接处理后的数据
data_movietime = pd.DataFrame(lists_end,index=data.index,columns=['电影时长(分钟)'])
datas = data.merge(data_movietime,left_on=data.index,right_on=data_movietime.index,how="outer")

# (6).修改电影时长(分钟)的数据类型
datas['电影时长(分钟)'] = datas['电影时长(分钟)'].astype('int')

# (7).将原片长和key_0列删除
columns = datas.columns[(datas.columns!='key_0') & (datas.columns!='片长')]
data = datas.loc[:,columns]

print()

# 查看处理后情况
data.info()
```

```python
# 8.电影类型处理

# 查看类型为空数据(数据虽然较少，但可以不用删除替换为其他)
data.dropna(subset=['类型']).info()

#将类型为nan替换为其他
data['类型'].fillna('其他',inplace=True)

print()

# 查看处理后情况
data.info()
```

```python
# 9. 电影评分处理

# 查看暂无电影评分数据(数据较少，可以删除)
data[data.loc[:,'评分']!='暂无电影评分'].info()

# 电影评分数据类型与小数点精确处理
data = data[data.loc[:,'评分']!='暂无电影评分']
data['评分'] = data['评分'].astype('float')
data['评分'] = data['评分'].fillna(data['评分'].mean()).round(1)

print()

# 查看处理后情况
data.info()
```

```python
# 10.电影票房处理

# 查看票房为空的电影信息(不能删除nan，否则没多少数据了)
data[pd.isnull(data['票房'])].info()

# 将票房的nan替换成票房最小值
data['票房'] = data['票房'].fillna(data['票房'].min())

print()

# 查看处理后情况
data.info()
```

```python
# 11.处理发行公司与制片公司(不能删除nan，否则没多少数据了)

# 查看相应公司为空数据
data.dropna(subset=['发行公司','制片公司']).info()

# 将nan值替换为暂无相应公司，也不能删除
data['发行公司'] = data['发行公司'].fillna('暂无发行公司')
data['制片公司'] = data['制片公司'].fillna('暂无制片公司')

print()

# 查看处理后情况
data.info()
```

```python
# 12.重新排序index和columns

# 重新排序行索引
indexs = []
for i in range(len(data)):
    indexs.append(i)
data.index = indexs

# 重新排序列索引
data = data.loc[:,['电影名','导演','编剧','演员','类型','电影时长(分钟)','上映时间(年)','评分','票房','发行公司','制片公司']]
```

```python
# 13.导出csv文件

# 准备两份数据
movies_all = data
movies_half = data.loc[:,['电影名','导演','编剧','演员','类型','电影时长(分钟)','上映时间(年)','评分']]

movies_all.to_csv("movies_all.csv",index=None,encoding = "utf-8") #包含豆瓣和艺恩数据
movies_half.to_csv("movies_half.csv",index=None,encoding = "utf-8") #只包含豆瓣数据
```
