```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
data = pd.read_csv('../2-Data_Cleaning/movies_all.csv')
```

```python
data
```

```python
data.info()
```

```python
# 1.数据选择与排序

#选取票房不为0的数据,并重新排序
data = data[data['票房']!=0]
list_index = [i for i in range(len(data))]
data.index = list_index

#查看重新排序后的数据
data
```

```python
# 2.构建演员矩阵

#演员列表嵌套解除嵌套并去重
list_Actor = data["演员"].str.split(",").tolist()
list_Actors = list(set([i for j in list_Actor for i in j]))

#构建演员为0的矩阵
zeros_data = pd.DataFrame(np.zeros((data.shape[0],len(list_Actors))),columns=list_Actors)

#给每个电影出现演员的位置加1,构建全新矩阵
for i in range(data.shape[0]):
    zeros_data.loc[i,list_Actor[i]] = 1

#查看演员矩阵    
zeros_data
```

```python
# 3.拼接演员矩阵和票房

#获取票房并转换为列表
BoxOffice = list(data['票房'])

#转换票房列表并拼接数据
data_BoxOffice = pd.DataFrame(BoxOffice,columns=['票房'])
data_zeros = pd.concat([zeros_data,data_BoxOffice],axis=1)

#查看全新数据
data_zeros
```

```python
# 4.拼接演员与总票房

#将演员的总票房算出并添加至列表
BoxOffice_All = []
for i in list_Actors:
    BoxOffice_All.append(data_zeros[data_zeros.loc[:,i]!=0]['票房'].sum().round(1))

#拼接列表为Dataframe
datas = pd.DataFrame({"演员":list_Actors,"总票房":BoxOffice_All})

#按票房大小逆序排序,取出总票房前10的编剧数据
Actors_ten = datas.sort_values(by='总票房',ascending=False).head(10)['演员'].values
BoxOffice_ten = datas.sort_values(by='总票房',ascending=False).head(10)['总票房'].values

#查看拼接数据
datas
```

```python
# 5.绘制图片

#设置图片大小
plt.figure(figsize=(12,9),dpi=80)

#设置字体
my_font = FontProperties(fname=r"C:\Windows\Fonts\simsun.ttc", size=12)
my_fonts = FontProperties(fname=r"C:\Windows\Fonts\方正粗黑宋简体.ttf", size=14)

#绘制图片
plt.bar(Actors_ten,BoxOffice_ten,width=0.8)

#设置x轴的刻度
plt.xticks(Actors_ten,Actors_ten,fontproperties=my_font)

#绘制表格
plt.grid(alpha=1) #alpha是表格透明度

#添加描述信息
plt.xlabel("演员",fontproperties=my_fonts,color='blue')
plt.ylabel("总票房(万)",fontproperties=my_fonts,color='green')
plt.title("总票房最高的10名演员",fontproperties=my_fonts,color='red')

#保存图片
plt.savefig("./总票房最高的10名演员.png") #保存为.svg格式矢量图，不会有失帧

#展示图片
plt.show()
```
