```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
data = pd.read_csv('../2-Data_Cleaning/movies_all.csv')
```

```python
data
```

```python
data.info()
```

```python
grouped = data.groupby(by='类型').count()
```

```python
grouped
```

```python
type(grouped['电影名'])
```

```python
type(grouped['电影名']/len(data))
```

```python
news = pd.concat([grouped['电影名'],grouped['电影名']/len(data)],axis=1)
```

```python
news.columns=['电影数目','电影占比']
```

```python
news['电影占比'] = news['电影占比'].apply(float)
```

```python
news
```

```python
type(news['电影占比'][0])
```

```python
news = news[news['电影占比'] > 0.001]
```

```python
x = list(news.index)
y = list(news['电影数目'].values)
```

```python
x
```

```python
y
```

```python
#设置图片大小
plt.figure(figsize=(16,12),dpi=80)

#用来正常显示中文标签
plt.rcParams['font.sans-serif']=['SimHei']

labels = x

sizes = y

plt.pie(sizes #每一块饼图的比例，为必填项
        ,labels=labels #每一块饼图外侧显示的说明文字
        ,autopct='%1.1f%%' #格式化
        ,shadow=False #阴影
        ,startangle=90 #起始绘制角度，默认图是从x轴正方向逆时针画起，如设定startangle=90则从y轴正方向画起
       ) 

#添加标签
plt.title("某月份家庭支出")

#保存图片
#plt.savefig("./Test.png",dpi=200,bbox_inches='tight')

#展示图片
plt.show()
```
