```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
data = pd.read_csv('../3-Data_Cleaning/movies_half.csv')
```

```python
data
```

```python
data.info()
```

```python
#计算组数
d = 10 #组距
time = data['电影时长(分钟)']
num_bins = (max(time)-min(time)) // d

#设置图片大小
plt.figure(figsize=(20,8),dpi=80)

#设置字体
my_font = FontProperties(fname=r"C:\Windows\Fonts\simsun.ttc", size=14)

#绘制图片
plt.hist(time,num_bins)

#设置x轴的刻度
plt.xticks(range(min(time),max(time)+d,d))

#绘制表格
plt.grid()

#添加描述信息
plt.xlabel("电影时长",fontproperties=my_font)
plt.ylabel("电影数",fontproperties=my_font)
plt.title("所有电影时长频数分布",fontproperties=my_font)

#保存图片
plt.savefig("./所有电影时长频数分布.png") #保存为.svg格式矢量图，不会有失帧

#展示图片
plt.show()
```
