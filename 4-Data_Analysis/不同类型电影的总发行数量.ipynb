```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
data = pd.read_csv('../3-Data_Cleaning/movies_all.csv')
```

```python
data
```

```python
data.info()
```

```python
# 1.数据分类

#电影类型分类
grouped = data.groupby(by='类型').count()

#数据拼接
news = pd.concat([grouped['电影名'],grouped['电影名']/len(data)],axis=1)
news.columns=['电影数目','电影占比']
news['电影占比'] = news['电影占比'].apply(float)
```

```python
# 2.选取数据

MovieType = list(news.index)
MovieNumber = list(news['电影数目'].values)
```

```python
# 3.绘制图片

#设置图片大小
plt.figure(figsize=(20,8),dpi=80)

#设置字体
my_font = FontProperties(fname=r"C:\Windows\Fonts\simsun.ttc", size=16)
my_fonts = FontProperties(fname=r"C:\Windows\Fonts\方正粗黑宋简体.ttf", size=20)

#绘制图片
plt.barh(range(len(MovieType)),MovieNumber,height=0.3,color="red")

#设置y轴的刻度
plt.yticks(range(len(MovieType)),MovieType,fontproperties=my_font)

#添加描述信息
plt.title("不同类型电影的总发行数量",fontproperties=my_fonts,color='red')
plt.xlabel("电影发行数量",fontproperties=my_font,color='blue')
plt.ylabel("电影类型",fontproperties=my_font,color='green')

#绘制表格
plt.grid()

#保存图片
plt.savefig("./不同类型电影的总发行数量.png") #保存为.svg格式矢量图，不会有失帧

#展示图片
plt.show()
```
