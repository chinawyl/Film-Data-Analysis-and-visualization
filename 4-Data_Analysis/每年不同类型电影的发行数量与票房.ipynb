```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
data = pd.read_csv('../3-Data_Cleaning/movies_all.csv')
```

```python
# 1.准备绘图数据

#添加电影数量列
data['电影数量'] = 1

#按照年份分类
grouped = data.groupby(by='上映时间(年)')

#获取时间列表并排序
time = data['上映时间(年)'].unique().tolist()
time.sort()
time = time[:-1]

#获取电影数量与票房(因为2020年才开始，故不做统计)
MovieNumber = grouped.sum()['电影数量'][:-1]
BoxOffice = grouped.sum()['票房'][:-1]
```

```python
# 2.绘制图片

fig,ax1=plt.subplots()

ax1.plot(MovieNumber,color='b',label='电影数量')
ax1.set_xlabel('年份')
ax1.set_ylabel('电影数量')
ax1.legend(loc='upper left')

ax2=ax1.twinx()
ax2.plot(BoxOffice,color='y',label='票房')
ax2.set_ylabel('票房')
ax2.legend(loc='upper right')

plt.savefig("./每年不同类型电影的发行数量与票房.png") #保存为.svg格式矢量图，不会有失帧

plt.show()
```
