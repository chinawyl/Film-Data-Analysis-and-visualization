```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
data = pd.read_csv('../3-Data_Cleaning/movies_all.csv')
```

```python
data
```

```python
data.info()
```

```python
# 1.数据选择与排序

#选取票房不为0的数据,并重新排序
data = data[data['票房']!=0]
list_index = [i for i in range(len(data))]
data.index = list_index

#查看重新排序后的数据
data
```

```python
# 2.绘制图片

#设置图片大小
plt.figure(figsize=(20,8),dpi=80)

#设置字体
my_font = FontProperties(fname=r"C:\Windows\Fonts\simsun.ttc", size=14)

#绘制图片
plt.scatter(data['评分'],data['票房'])

#添加描述信息
plt.xlabel("评分",fontproperties=my_font)
plt.ylabel("票房",fontproperties=my_font)
plt.title("票房与评分的关系",fontproperties=my_font)

#保存图片
plt.savefig("./票房与评分的关系.png") #保存为.svg格式矢量图，不会有失帧

#展示图片
plt.show()
```
