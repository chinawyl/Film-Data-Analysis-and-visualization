```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python
data = pd.read_csv('../3-Data_Cleaning/movies_all.csv')
```

```python
data.info()
```

```python
# 1.数据分类

#电影类型分类
grouped = data.groupby(by='类型').count()

#数据拼接
news = pd.concat([grouped['电影名'],grouped['电影名']/len(data)],axis=1)
news.columns=['电影数目','电影占比']
news['电影占比'] = news['电影占比'].apply(float)
```

```python
# 2.选取数据

#选择占比大于0.001的数据，不然饼图画出来要重叠
news = news[news['电影占比'] > 0.001]

#选择饼图的labels与sizes
labels = list(news.index)
sizes = list(news['电影数目'].values)
```

```python
# 3.绘制图片

#设置图片大小
plt.figure(figsize=(16,12),dpi=80)

#用来正常显示中文标签
plt.rcParams['font.sans-serif']=['SimHei']

labels = labels

sizes = sizes

plt.pie(sizes #每一块饼图的比例，为必填项
        ,labels=labels #每一块饼图外侧显示的说明文字
        ,autopct='%1.1f%%' #格式化
        ,shadow=False #阴影
        ,startangle=90 #起始绘制角度，默认图是从x轴正方向逆时针画起，如设定startangle=90则从y轴正方向画起
       ) 

#添加标签
plt.title("不同类型电影的总发行数量占比")

#保存图片
plt.savefig("./不同类型电影的总发行数量占比.png",dpi=200,bbox_inches='tight')

#展示图片
plt.show()
```
