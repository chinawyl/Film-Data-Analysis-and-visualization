```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
data = pd.read_csv('../3-Data_Cleaning/movies_half.csv')
```

```python
data
```

```python
data.info()
```

```python
# 1.数据分类 

#添加列评分次数
data['评分次数'] = 1

#按电影类型分类并计算平均评分
grouped = data.groupby(by='类型')
datas = grouped.sum()
datas['平均评分'] = datas['评分']/datas['评分次数']

#选择绘图列
MovieType = list(datas.index)
AverageScore = list(datas['平均评分'])
AverageScore.sort()
```

```python
# 2.绘制图片

#设置图片大小
plt.figure(figsize=(20,8),dpi=80)

#设置字体
my_font = FontProperties(fname=r"C:\Windows\Fonts\simsun.ttc", size=16)
my_fonts = FontProperties(fname=r"C:\Windows\Fonts\simsun.ttc", size=20)

#绘制图片
plt.barh(range(len(MovieType)),AverageScore,height=0.3,color="orange")

#设置y轴的刻度
plt.yticks(range(len(MovieType)),MovieType,fontproperties=my_font)

#添加描述信息
plt.xlabel("平均评分",fontproperties=my_font,color='blue')
plt.ylabel("电影类型",fontproperties=my_font,color='purple')
plt.title("不同类型电影的平均评分",fontproperties=my_fonts,color='red')


#绘制表格
plt.grid()

#保存图片
plt.savefig("./不同类型电影的平均评分.png") #保存为.svg格式矢量图，不会有失帧

#展示图片
plt.show()
```
