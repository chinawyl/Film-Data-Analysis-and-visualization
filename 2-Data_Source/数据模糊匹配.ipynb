```python
import numpy as np
import pandas as pd
```

```python
# 1.合并数据

# 豆瓣数据
data_douban = pd.read_csv('./douban.csv')
data_douban.drop_duplicates(subset=['电影名'],inplace=True)

# 合并艺恩数据
yien1 = pd.read_csv('./yien1.csv')  
yien2 = pd.read_csv('./yien2.csv') 
data_yien = pd.concat([yien1,yien2])

# 处理艺恩票房数据
data_yien = data_yien.loc[:,data_yien.columns != '_id']
data_yien = data_yien.loc[:,['MovieName','SumBoxOffice','MovieFxAll','MovieZz']]
data_yien.columns = ['电影名','票房','发行公司','制片公司']
data_yien.drop_duplicates(subset=['电影名'],inplace=True)

# 合并处理后的数据
data_final = pd.read_csv('../3-Data_Cleaning/movies_all.csv')
```
